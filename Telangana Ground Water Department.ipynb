```python
import pandas as pd
df_2018_safe = df_2018[df_2018['TDS_Status'] == 'Safe'][['TDS', 'district']]
df_2019_safe = df_2019[df_2019['TDS_Status'] == 'Safe'][['TDS', 'district']]
df_2020_safe = df_2020[df_2020['TDS_Status'] == 'Safe'][['TDS', 'district']]
df_2021_safe = df_2021[df_2021['TDS_Status'] == 'Safe'][['TDS', 'district']]
df_2022_safe = df_2022[df_2022['TDS_Status'] == 'Safe'][['TDS', 'district']]

# Concatenate dataframes for each year
all_data_safe = pd.concat([df_2018_safe, df_2019_safe, df_2020_safe, df_2021_safe, df_2022_safe])
suffixes=('_2019', '_2020')
# Display the concatenated dataframe
print(all_data_safe)
```

```python
import pandas as pd

# Selecting 'TDS' and 'District' columns from each dataframe and filtering rows where 'TDS_Status' is 'Safe'
df_2018_safe = df_2018[df_2018['TDS_Status'] == 'Safe'][['TDS', 'district']]
df_2019_safe = df_2019[df_2019['TDS_Status'] == 'Safe'][['TDS', 'district']]
df_2020_safe = df_2020[df_2020['TDS_Status'] == 'Safe'][['TDS', 'district']]
df_2021_safe = df_2021[df_2021['TDS_Status'] == 'Safe'][['TDS', 'district']]
df_2022_safe = df_2022[df_2022['TDS_Status'] == 'Safe'][['TDS', 'district']]

# Concatenate dataframes for each year with suffixes
all_data_safe = pd.concat([df_2018_safe, df_2019_safe, df_2020_safe, df_2021_safe, df_2022_safe], 
                          keys=['2018', '2019', '2020', '2021', '2022'], 
                          names=['Year'])

# Display the concatenated dataframe
print(all_data_safe)
```

```python
all_data_safe
```

```python
import matplotlib.pyplot as plt


count_per_year = all_data_safe.groupby(level='Year').size()


plt.figure(figsize=(10, 6))
count_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Districts with Safe Underground Water per Year')
plt.xlabel('Year')
plt.ylabel('Number of Districts')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each district across all years
district_occurrences = all_data_safe['district'].value_counts()

# Find the districts with the maximum occurrences
max_occurrence_districts = district_occurrences[district_occurrences == district_occurrences.max()]

# Display the districts with the maximum occurrences
print(max_occurrence_districts)
```

```python
all_data_safe = all_data_safe.reset_index()
```

```python
# Iterate over each district in max_occurrence_districts
for district in max_occurrence_districts.index:
    # Find the years in which the district occurs
    years = all_data_safe.loc[all_data_safe['district'] == district, 'Year'].unique()
    
    # Display the district and the years of occurrence
    print(f"District: {district}, Years: {', '.join(map(str, years))}")
```
